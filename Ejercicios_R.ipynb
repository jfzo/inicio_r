## Listas

Recuerde que una lista corresponde a una colección heterogenea de valores, i.e. distintos tipos de datos pueden ser recolectados. Para construir una lista se utiliza el comando _list()_, cuyos parámetros corresponden a los elementos que inicialmente constituiran la colección. Por ejemplo:
```R
        L <- list(4, c(21,33), "Ausente")
```
corresponde a una lista con 3 elementos (número, vector y chr. Puede ser comprobado con el comando _str_).
Todos los elementos son ubicados secuencialmente en esta colección y a cada uno se le asigna un indice (al igual que con los vectores). Por lo tanto para acceder a un elemento se requiere únicamente el índice asociado a su posición:
```R
        L[[2]]
```
En este caso se obtendrá el vector almancenado en la segunda posición de la lista simbolizada por la variable _x_. Notar que para el caso de las listas, se usa el doble corchete, ya que el uso del corchete simple (Por ejemplo $L[2]$)  entrega una lista con las posiciones contenidas. El uso de esta última notación permite acceder a varios elementos a la vez, por ejemplo para obtener los últimos dos elementos de la lista $L$ se utiliza la expresión:
```R
        L[c(2,3)]
```
Para agregar elementos a una lista, basta con colocar la posición en la que se agregará el nuevo elemento y asignarle el valor deseado. Por ejemplo, para agregar un nuevo vector a la lista $L$ que tiene largo 3 actualmente se realiza lo siguiente:
```R
        L[[4]] <- c(1,6,29)
```
Puede verificarse que la lista cambió con el comando _str_ o simplemente con _length_.
